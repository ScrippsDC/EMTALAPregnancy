```python
rg_manual = pandas.read_excel("../data/manual/EMTALA_PREGNANT_2011-2022.xlsx")
rg_manual["inspection_date_dt"] = pandas.to_datetime(rg_manual["inspection_date"],errors="coerce")

print(rg_manual[rg_manual['inspection_date_dt']<=DOBBS]["facility_id"].nunique())
print(rg_manual[rg_manual['inspection_date_dt']>DOBBS]["facility_id"].nunique())
```

```python
# Filter rg_manual for only facilities with a number -- as opposed to a string -- in Unnamed: 0
rg_manual_counted = rg_manual[rg_manual["Unnamed: 0"].apply(lambda x: (isinstance(x, int))|(isinstance(x, float))) & ~rg_manual["Unnamed: 0"].isna()]
rg_manual_counted["facility_id"].nunique()

rg_manual_counted[~(rg_manual_counted["inspection_date_dt"]<="01-01-2025")].head().transpose()
```
